### Quais as restrições?

#### Restrições tecnicas e Restrições de negatividade


```python
@latexify.with_latex
def restricoes(x1, x2, x3):
  if 1:
    return (x1 >= 3)
  elif 2:
    return (3*x1 + 4*x2 >= 30)
  elif 3:
    return (3*x1 + 10*x3 >= 30) 
  elif 4:
    return (x1 + x2 + x3 >= 10)
  elif 5:
    return x1 >= 0  
  elif 6: 
    return x2 >= 0
  else:
    return x3 >= 0    
    

restricoes  
```

# Programação

```python
# Definir o modelo
model = pulp.LpProblem('Lista 1 - Exercicio 8', sense=pulp.LpMaximize)

# Adicionar as variáveis
x = pulp.LpVariable.dicts(indices=[1, 2, 3], cat=pulp.LpContinuous, lowBound=0, name='x')

# Adicionar as Restruições
model.addConstraint(x[1] >= 3, name='restricao_1')
model.addConstraint(3*x[1] + 4*x[2] >= 30, name='restricao_2')
model.addConstraint(3*x[1] + 10*x[3] >= 30, name='restricao_3')
model.addConstraint(x[1] + x[2] + x[3] <= 10, name='restricao_4')

# Função Objetivo
model.setObjective(1000*x[1] + 1000*x[2] + 1000*x[3])

# Optimizar
model.solve()

# Obter e imprimir a solução
x_sol = {i: x[i].value() for i in [1, 2, 3]}
print(f'x = {x_sol}')
z = 1000*x_sol[1] + 1000*x_sol[2] + 1000*x_sol[3]
print(f'z = {z}')
```
