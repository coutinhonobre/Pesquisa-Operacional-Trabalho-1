```python
@latexify.with_latex
def melhorRetorno(x1, x2, x3):
  return 300*x1 + 400*x2 + 500*x3
melhorRetorno 
```

### Quais as restrições?

#### Restrições tecnicas e Restrições de negatividade

```python
@latexify.with_latex
def restricoes(x1, x2, x3):
  if 1:
    return ((x1 + x2 + x3) <= 100)
  elif 2:
    return (100000*x1 + 200000*x2 <= 12750000)
  elif 3:
    return (100*x1 + 200*x2 <= 14000) 
  elif 4:
    return x1 >= 0
  elif 5: 
    return x2 >= 0
  else:
    return x3 >= 0    
    

restricoes  
```

Observação: `if n` foi usado para indicar que é uma restrição, sendo `if` neste foi usado para indicar qual a restrição e `n` para indica a restrição.
O `otherwise` está sendo usado para indicar restrição de não negatividade
####**Restrições** 


*   Área total=> `if 1`
*   Água => `if 2`
*   Adubo => `if 3`
*   Restrição de não negatividade => `if 4`
*   Restrição de não negatividade => `if 5`
*   Restrição de não negatividade => `otherwise`

# Programação

```python
# Definir o modelo
model = pulp.LpProblem('Lista 1 - Exercicio 7', sense=pulp.LpMaximize)

# Adicionar as variáveis
x = pulp.LpVariable.dicts(indices=[1, 2, 3], cat=pulp.LpContinuous, lowBound=0, name='x')

# Adicionar as Restruições
model.addConstraint(x[1] + x[2] + x[3] <= 100, name='restricao_1')
model.addConstraint(100*x[2] + 200*x[3] <= 14000, name='restricao_2')
model.addConstraint(100000*x[2] + 200000*x[3] <= 12750000, name='restricao_3')

# Função Objetivo
model.setObjective(300*x[1] + 400*x[2] + 500*x[3])

# Optimizar
model.solve()

# Obter e imprimir a solução
x_sol = {i: x[i].value() for i in [1, 2, 3]}
print(f'x = {x_sol}')
z = 300*x_sol[1] + 400*x_sol[2] + 500*x_sol[3]
print(f'z = {z}')
```
